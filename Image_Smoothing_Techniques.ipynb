###**Robert's Operator**

```python
import cv2
import numpy as np

image = cv2.imread('/content/messenger.png', cv2.IMREAD_GRAYSCALE)

# Define Robert's operator kernels
roberts_x = np.array([[1, 0],
                      [0, -1]])

roberts_y = np.array([[0, 1],
                      [-1, 0]])

# Apply the convolution with the kernels
gradient_x = cv2.filter2D(image, cv2.CV_64F, roberts_x)
gradient_y = cv2.filter2D(image, cv2.CV_64F, roberts_y)

# Calculate the gradient magnitude
#gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
gradient_magnitude=cv2.magnitude(gradient_x,gradient_y)
# Normalize the gradient magnitude to the range [0, 255]
#gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

# Display the original image and the edge-detected image
#cv2.imshow('Original Image', image)
#cv2.imshow('Edge Detected (Robert\'s Operator)', gradient_magnitude)
#cv2.imshow('Original Image', image)
cv2.imwrite('robert.png', gradient_magnitude)
"""cv2.waitKey(0)
cv2.destroyAllWindows()"""
```

###**Sobel Operator**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/messenger.png', cv2.IMREAD_GRAYSCALE)

# Apply the Sobel operator
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Sobel horizontal
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Sobel vertical

# Calculate the magnitude of the gradient
magnitude = cv2.magnitude(sobelx, sobely)

# Normalize the magnitude to the range [0, 255]
#magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)

# Convert to uint8 (8-bit) format
#magnitude = np.uint8(magnitude)

# Display the original and edge-detected images
#cv2.imshow('Original Image', image)
#cv2.imshow('Sobel Edge Detection', magnitude)
cv2.imwrite("Sobel.png",magnitude)
"""cv2.waitKey(0)
cv2.destroyAllWindows()"""
```

###**Prewitt Operator**

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Load the image (replace 'image.jpg' with your image file)
image = cv2.imread('/content/messenger.png', cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Prewitt operator kernels
prewitt_x = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])

prewitt_y = np.array([[-1, -1, -1],
                      [0, 0, 0],
                      [1, 1, 1]])
```

```python
# Convolve the image with the Prewitt kernels
gradient_x = cv2.filter2D(image, -1, prewitt_x)
gradient_y = cv2.filter2D(image, -1, prewitt_y)
```

```python
# Calculate the magnitude of the gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
```

```python
# Display the gradient magnitude image
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(gradient_x, cmap='gray')
plt.title('Gradient in X direction')
plt.axis('off')

plt.subplot(132)
plt.imshow(gradient_y, cmap='gray')
plt.title('Gradient in Y direction')
plt.axis('off')

plt.subplot(133)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude')
plt.axis('off')

plt.show()
```

###**Canny Edge Detection**

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('/content/messenger.png', cv2.IMREAD_GRAYSCALE)

# Step 1: Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Step 2: Apply Canny Edge Detection
edges = cv2.Canny(blurred, threshold1=30, threshold2=100)

# Step 3: Optional - Further post-processing or visualization
# You can perform additional processing or display the edges
# For example, you can save the edges to an image or display them:
cv2.imwrite('edges.jpg', edges)

# Display the original image and the edges (for visualization)
#cv2.imshow('Original Image', image)
#cv2.imshow('Canny Edges', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
